**1. Peano axiomatization of integers**

According to G. Peano, to generate all natural integers one needs three ingredients 

1.   the number zero
2.   the number one
3.   a recursive function called a successor 

Giuseppe Peano (1858â€“1932), Italian mathematician.


```python
def successor(n):
  if n ==0:
    return 1
  else:
    return successor(n-1) +1
```

```python
successor(0)
```

```python
successor(1)
```

```python
successor(2)
```

**Theorem 1.:** Integers are infinite.

**Proof:** To prove that there is no largest number, take the so called 'biggest' number, and add one, or any number for that matter, to it, creating an even larger number than before.

**2. Recursive Functions**

A recursive function is a function that calls itself within the code. This function stops when it reaches a certain point, or a base case. 


Below are some examples of such functions

```python
def factorial(n):
  if n == 0:
    return 1
  else:
    return n*factorial(n-1)
```

```python
 factorial(200)
```

```python
def fibonnaci(n):
  if n == 0:
    return 0
  elif n == 1:
    return 1
  else:
    return fibonnaci(n-1) + fibonnaci(n-2)
```

```python
fibonnaci(10)
```

**3. Prime numbers.**


**Definition:** A *prime number* is an integer, bigger than one, that can only be divided by one and itself. Examples: 2, 3, 5 and 7.


Here is a function that tests whether an integer is a prime number or not. 

```python
def is_prime(n, verbose=False):
  for i in range(2, n):
    if n % i == 0:
      if verbose:
        print (f'divisor {i} found for {n}')
      return False
  return True
    
```

```python
is_prime(9, verbose=True)
```

**Theorem 2**. The set of prime numbers is infinite.

**Proof:** Multiply all known prime integers and add one. This causes you to always have a non-zero remainder.
